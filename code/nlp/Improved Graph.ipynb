```python
import json
import random
import configparser
import matplotlib.pyplot as plt
import networkx as nx
import spacy
import umap
from IPython.display import clear_output
from spacy.training.example import Example
from spacy.scorer import Scorer
from spacy.util import minibatch, compounding
from sklearn.model_selection import KFold
from spacy.training import offsets_to_biluo_tags
from spacy.util import compile_infix_regex
```

```python

def load_config(config_path):
    config = configparser.ConfigParser()
    config.read(config_path)
    return config

def create_graph_from_data(data):
    G = nx.DiGraph()

    all_entities = [entity for sample in data['samples'] for entity in sample['annotation']['entities']]
    all_relations = [relation for sample in data['samples'] for relation in sample['annotation']['relations']]

    for entity in all_entities:
        G.add_node(entity['textId'], name=entity['text'], nodetype=entity['label'])

    for relation in all_relations:
        G.add_edge(relation['from'], relation['to'], label=relation['label'])

    return G

def draw_graph(G, layout=None):
    def filter_nodes():
        return [n for n in G.nodes if G.nodes[n]["nodetype"] != "na"]

    def compute_spring_layout():
        return nx.spring_layout(filtered_subgraph, seed=42)

    def compute_umap_layout():
        embedding = umap.UMAP(n_neighbors=15).fit_transform(nx.to_numpy_array(filtered_subgraph))
        return {n: embedding[i] for i, n in enumerate(filtered_subgraph)}

    def draw_default_graph():
        fig, ax = plt.subplots()
        nx.draw(G.subgraph(filtered_nodes), with_labels=True, node_size=50, node_color="skyblue", alpha=0.5, linewidths=40)
        nx.draw_networkx_edge_labels(G, nx.spring_layout(G), edge_labels={(u, v): d["label"] for u, v, d in G.edges(data=True) if G.nodes[u]["nodetype"] != "na" and G.nodes[v]["nodetype"] != "na"}, font_size=12)

    def draw_custom_graph(pos):
        fig, ax = plt.subplots()
        nx.draw(G.subgraph(filtered_nodes), pos, labels={n: G.nodes[n]['name'] for n in filtered_nodes}, font_size=12, node_size=50, node_color="skyblue", alpha=0.5, linewidths=40)
        nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d["label"] for u, v, d in G.edges(data=True) if G.nodes[u]["nodetype"] != "na" and G.nodes[v]["nodetype"] != "na"}, font_size=12)

    filtered_nodes = filter_nodes()
    filtered_subgraph = G.subgraph(filtered_nodes)

    if layout == 'spring':
        pos = compute_spring_layout()
        draw_custom_graph(pos)
    elif layout == 'umap':
        pos = compute_umap_layout()
        draw_custom_graph(pos)
    else:
        draw_default_graph()

    plt.show()
    

def create_train_data(graph_data):
    entity_train_data = []
    relation_train_data = []

    for sample in data['samples']:
        document = sample["document"]
        annotation = sample["annotation"]
        
        # Entity data
        entities_data = [(entity["start"], entity["end"] + 1, entity["label"]) for entity in annotation["entities"]]
        entity_train_data.append((document, {"entities": entities_data}))
        
        # Relation data
        relations_data = []
        for relation in annotation['relations']:
            from_entity = next(ent for ent in annotation['entities'] if ent['textId'] == relation['from'])
            to_entity = next(ent for ent in annotation['entities'] if ent['textId'] == relation['to'])
            relations_data.append((relation['label'], (from_entity['start'], from_entity['end'] + 1), (to_entity['start'], to_entity['end'] + 1)))
        relation_train_data.append((document, relations_data))

    return entity_train_data, relation_train_data

def evaluate_entity_model(nlp, test_data):
    scorer = Scorer(nlp)
    for example in test_data:
        pred_doc = nlp(example.text)
        ref_doc = nlp(example.reference.text)
        ref_doc.ents = example.reference.ents

        eval_example = Example.from_dict(pred_doc, {"entities": [(ent.start_char, ent.end_char, ent.label_) for ent in ref_doc.ents]})
        scores = scorer.score([eval_example])

    return scores

# Load the JSON data
with open(r"H:\Curated-LLM-Data\Universal Data Tool\New text_entity_recognition Dataset.udt.json", 'r') as f:
    data = json.load(f)

# Create the directed graph
G = create_graph_from_data(data)

# Output JSON with all nodes and relations
graph_data = {
    "nodes": [],
    "edges": []
}

for node in G.nodes:
    graph_data["nodes"].append({
        "id": node,
        "name": G.nodes[node]["name"],
        "nodetype": G.nodes[node]["nodetype"]
    })

for edge in G.edges:
    graph_data["edges"].append({
        "from": edge[0],
        "to": edge[1],
        "label": G.edges[edge]["label"]
    })

with open("graph.json", "w") as f:
    json.dump(graph_data, f, indent=2)
    
if(False):
    for n in G.nodes:
        print(G.nodes[n]["nodetype"])
        
# Load the graph data
with open("graph.json", "r") as f:
    graph_data = json.load(f)

# Compute UMAP layout
#pos = compute_umap_positions(G)

# Create train_data
entity_train_data, relation_train_data = create_train_data(data)
train_data = entity_train_data

def preprocess_relation_data(data):
    preprocessed_data = []
    
    for document, relations in data:
        doc = nlp(document)
        entity_pairs = [(ent1, ent2) for ent1 in doc.ents for ent2 in doc.ents if ent1 != ent2 and ent1.label_ != 'na' and ent2.label_ != 'na']
        
        for relation in relations:
            label, (start1, end1), (start2, end2) = relation
            for ent1, ent2 in entity_pairs:
                if ent1.start_char == start1 and ent1.end_char == end1 and ent2.start_char == start2 and ent2.end_char == end2:
                    example_text = doc[ent1.start:ent2.end].text
                    preprocessed_data.append((example_text, {"relations": [(label, (0, len(ent1.text)), (len(example_text) - len(ent2.text), len(example_text)))]}))
                    
    return preprocessed_data

# Load the trained NER model
nlp = spacy.load("trained_ner_model")
    
# Extract entities using the trained NER model

###start diagnostics before preprocess_relation_data

for i, (doc, relations) in enumerate(relation_train_data[:3]):
    print(f"Example {i+1}:")
    print(f"Document: {doc}")
    pred_doc = nlp(doc)
    print(f"Entities: {[(ent.start_char, ent.end_char, ent.text) for ent in pred_doc.ents]}")
    print()
###end diagnostics

preprocessed_relation_data = preprocess_relation_data(relation_train_data)

#validate
# Print a few examples from relation_train_data
print("Examples from relation_train_data:")
for i, (doc, relations) in enumerate(relation_train_data[:3]):
    print(f"Example {i+1}:")
    print(f"Document: {doc}")
    print(f"Relations: {relations}")
    print()

# Print a few examples from preprocessed_relation_data
print("Examples from preprocessed_relation_data:")
for i, (example_text, example_relations) in enumerate(preprocessed_relation_data[:3]):
    print(f"Example {i+1}:")
    print(f"Example Text: {example_text}")
    print(f"Example Relations: {example_relations}")
    print()

# Load the config file
config_path = 'config.ini'
config = load_config(config_path)

# Create a blank spaCy NER model
nlp = spacy.blank("en")
ner = nlp.add_pipe("ner")

# Add the labels to the model
for label in set(entity["nodetype"] for entity in graph_data["nodes"]):
    ner.add_label(label)

# Set a threshold for the loss value
loss_threshold = 0.05

# Training settings
n_splits_outer = 2
n_splits_inner = 2
epochs = 200

kf_outer = KFold(n_splits=n_splits_outer, shuffle=True, random_state=42)
kf_inner = KFold(n_splits=n_splits_inner, shuffle=True, random_state=42)

train_examples = [Example.from_dict(nlp.make_doc(text), ann) for text, ann in train_data]

# Initialize the plot
plt.ion()
fig_loss, ax_loss = plt.subplots()

# Pre-allocate the x-axis and y-axis data
x = list(range(1, epochs + 1))
y = [0] * epochs

optimizer = nlp.begin_training()
optimizer.learn_rate = float(config["training.optimizer"]["learn_rate"])
optimizer.b1 = float(config["training.optimizer"]["beta1"])
optimizer.b2 = float(config["training.optimizer"]["beta2"])
optimizer.L2 = float(config["training.optimizer"]["L2"])
optimizer.eps = float(config["training.optimizer"]["eps"])
optimizer.L2_is_weight_decay = config["training.optimizer"]["L2_is_weight_decay"]

for epoch in range(epochs):
    outer_scores = []

    for train_index_outer, test_index_outer in kf_outer.split(train_examples):
        train_data_outer = [train_examples[i] for i in train_index_outer]
        test_data_outer = [train_examples[i] for i in test_index_outer]

        for train_index_inner, valid_index_inner in kf_inner.split(train_data_outer):
            train_data_inner = [train_data_outer[i] for i in train_index_inner]
            valid_data_inner = [train_data_outer[i] for i in valid_index_inner]

            random.shuffle(train_data_inner)
            losses = {}
            batches = minibatch(train_data_inner, size=compounding(4.0, 32.0, 1.001))
            for batch in batches:
                nlp.update(batch, sgd=optimizer, drop=0.15, losses=losses)

            scores_inner = evaluate_entity_model(nlp, valid_data_inner)

        scores_outer = evaluate_entity_model(nlp, test_data_outer)
        outer_scores.append(scores_outer['ents_f'])

    mean_outer_score = sum(outer_scores) / len(outer_scores)

    y[epoch] = mean_outer_score
    clear_output(wait=True)
    plt.plot(x[:epoch + 1], y[:epoch + 1])
    plt.xlabel("Epoch")
    plt.ylabel("Mean Outer F1-score")
    plt.show()

    if mean_outer_score > (1 - loss_threshold):
        print("Making predictions on the last test data.")
        for test_example in test_data_outer:
            sample_text = test_example.text
            doc = nlp(sample_text)
            print(f"Entities in '{sample_text}'")
            for ent in doc.ents:
                print(ent.label_, ent.text)

        print(f"Training stopped at epoch {epoch + 1} with F1-score {mean_outer_score}, above threshold {1 - loss_threshold}")
        break

# Keep the plot open after training is finished
plt.ioff()
plt.show()

# Save the model
nlp.to_disk("trained_ner_model")

# Load the saved model
nlp = spacy.load("trained_ner_model")
    
draw_graph(G,layout='umap')
draw_graph(G,layout='spring')
```

```python

#Now, train the combined model using K-Fold Cross-Validation:



def evaluate_combined_model(nlp, test_data):
    textcat = nlp.get_pipe("textcat_multilabel")
    entity_scorer = Scorer(nlp)
    cat_scorer = Scorer(nlp)

    for example in test_data:
        pred_doc = nlp(example.x.text)
        ref_relations = [label for label, value in example.y.cats.items() if value]

        # Check for misaligned entities before creating the evaluation example
        ent_offsets = [(ent.start_char, ent.end_char, ent.label_) for ent in example.reference.ents]
        biluo_tags = offsets_to_biluo_tags(nlp.make_doc(example.x.text), ent_offsets)
        misaligned_entities = [offsets for tag, offsets in zip(biluo_tags, ent_offsets) if tag == '-']
        if misaligned_entities:
            print(f"Misaligned entities in '{example.x.text}': {misaligned_entities}")

        eval_example = Example.from_dict(pred_doc, {"entities": ent_offsets, "cats": {label: 1.0 if label in ref_relations else 0.0 for label in textcat.labels}})
        entity_scores = entity_scorer.score([eval_example])
        cat_scores = cat_scorer.score([eval_example])

    mean_score = (entity_scores['ents_f'] + cat_scores['cats_macro_f']) / 2
    return mean_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)
optimizer = nlp.begin_training()
epochs = 10

for epoch in range(epochs):
    scores = []
    for train_index, test_index in kf.split(combined_examples):
        train_data = [combined_examples[i] for i in train_index]
        test_data = [combined_examples[i] for i in test_index]

        # Shuffle the training data
        random.shuffle(train_data)

        # Update the model
        losses = {}
        batches = minibatch(train_data, size=compounding(4.0, 32.0, 1.001))
        for batch in batches:

            # Check for misaligned entities before updating the model
            for example in batch:
                ent_offsets = [(ent.start_char, ent.end_char, ent.label_) for ent in example.reference.ents]
                biluo_tags = offsets_to_biluo_tags(nlp.make_doc(example.x.text), ent_offsets)
                misaligned_entities = [offsets for tag, offsets in zip(biluo_tags, ent_offsets) if tag == '-']
                if misaligned_entities:
                    print(f"Misaligned entities in '{example.x.text}': {misaligned_entities}")

            nlp.update(batch, sgd=optimizer, drop=0.15, losses=losses)

        # Evaluate the model         
        mean_score = evaluate_combined_model(nlp, test_data)
        scores.append(mean_score)

    # Calculate the mean F1-score for this epoch
    mean_score = sum(scores) / len(scores)
    print(f"Epoch {epoch + 1} Mean F1-score: {mean_score}")
    
# Save the combined model
nlp.to_disk("trained_combined_model")

# Load the saved combined model
nlp = spacy.load("trained_combined_model")

#Now, you can use the combined model to predict both entities and relations for new data. To do this, first create a function to predict entities using the NER component:

def predict_entities(nlp, text):
    doc = nlp(text)
    entities = [(ent.start_char, ent.end_char, ent.label_) for ent in doc.ents]
    return entities

#Next, create a function to predict relations using the TextCategorizer component:

def predict_relations(nlp, text):
    doc = nlp(text)
    entities = [(ent.start_char, ent.end_char, ent.label_, ent.text) for ent in doc.ents if ent.label_ != "na"]
    
    relations = []
    for i in range(len(entities)):
        for j in range(i+1, len(entities)):
            ent1_start, ent1_end, ent1_label, ent1_text = entities[i]
            ent2_start, ent2_end, ent2_label, ent2_text = entities[j]
            
            pair_text = text[ent1_start:ent2_end]
            pair_doc = nlp(pair_text)
            
            for label, score in pair_doc.cats.items():
                if score > 0.5:
                    relations.append((label, ent1_text, ent2_text, score))
    
    return relations

#Now, you can use these functions to predict entities and relations fornew data:

# Sample text to predict entities and relations
sample_text = "John works at Google."

# Predict entities
entities = predict_entities(nlp, sample_text)
print(f"Entities in '{sample_text}':")
for start, end, label in entities:
    print(label, sample_text[start:end])

# Predict relations
relations = predict_relations(nlp, sample_text)
print(f"Relations in '{sample_text}':")
for label, ent1_text, ent2_text, score in relations:
    print(label, ent1_text, ent2_text, score)
```
