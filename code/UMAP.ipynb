```python
#!pip install umap-learn pandas numpy matplotlib
import pandas as pd
import numpy as np
import umap.umap_ as umap
import matplotlib.pyplot as plt

# Create a sample pandas dataframe
df = pd.read_csv('https://raw.githubusercontent.com/thistleknot/Python-Stock/master/data/raw/states.csv?token=GHSAT0AAAAAAB5KW4VMBY2VWS4UHDQ32WX2ZA6KEVA').set_index('States')

# Apply UMAP and reduce the dataframe to a 2-dimensional matrix
reducer = umap.UMAP(n_neighbors=15, n_components=2, random_state=42)
embedding = reducer.fit_transform(df)

# Plot the 2D UMAP projection using matplotlib
fig, ax = plt.subplots(figsize=(10, 8))
scatter = ax.scatter(embedding[:, 0], embedding[:, 1], s=100)

# Add index names to the plot
for i, txt in enumerate(df.index):
    ax.annotate(txt, (embedding[i, 0], embedding[i, 1]), fontsize=12, fontweight='bold')

# Set plot labels and title
ax.set_xlabel('UMAP Dimension 1', fontsize=14)
ax.set_ylabel('UMAP Dimension 2', fontsize=14)
ax.set_title('2D UMAP Projection of the DataFrame', fontsize=16)

# Show the plot
plt.show()
```
